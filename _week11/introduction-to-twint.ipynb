```python
# Import libraries
import nest_asyncio
import twint



```

```python
nest_asyncio.apply()
# Configure
c = twint.Config()
# Let's look at the tweets of Jorts the Cat
c.Username = "JortsTheCat"
# Let's search for all tweets that contain the string 'paw'
c.Search = "paw"
c.Limit = 10
c.Since = "2022-01-27"
# Save our output as a CSV file
c.Output = "./jorts-paw-tweets.csv"
c.Store_csv = True


# Run
twint.run.Search(c)
```
